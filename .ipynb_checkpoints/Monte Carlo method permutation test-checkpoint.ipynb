```python
import numpy as np
import pandas as pd

import plotly.graph_objects as go
```

- Permutation test samples w/o replacement unlike bootstraping

```python
# No difference
a_n, b_n = 10000, 1000 
a_group = np.random.uniform(low=0, high=100, size=(a_n,))
b_group = np.random.uniform(low=5, high=100, size=(b_n,))

true_ctr_diff = a_group.mean() - b_group.mean()
```

```python
r = 10000
combined = np.concatenate([a_group, b_group]) 
ctr_diff_list = list()
for i in range(r):
    np.random.shuffle(combined)
    a_sample = combined[:a_n]
    b_sample = combined[a_n:]
    
    ctr_diff = a_sample.mean() - b_sample.mean()
    ctr_diff_list.append(ctr_diff)
```

```python
fig = go.Figure()

fig.add_trace(
    go.Histogram(
        x=ctr_diff_list
    )
)

fig.add_vline(x=true_ctr_diff, line_width=3, line_dash="dash", line_color="green",
             annotation_text="true_ctr_diff")

fig.update_layout(
    title = f"permuation test with {r} iteration",
    xaxis_title="CTR (%)",
    yaxis_title="freq"
)

fig.show()
```

```python
mean = np.mean(ctr_diff_list)
std_dev = np.std(ctr_diff_list)

print("True ctr diff = ", true_ctr_diff)
print([mean - 2*std_dev, mean + 2*std_dev])
```
